```python
import React from "react";
import { motion } from "framer-motion";

// GoalDrivenDashboard.jsx
// Single-file React component using Tailwind CSS classes.
// Default export is the dashboard component. Replace mock data with real API data as needed.

export default function GoalDrivenDashboard() {
  // Mock data
  const balance = 12450;
  const goals = [
    { id: 1, title: "Emergency Fund", current: 720, target: 1000 },
    { id: 2, title: "Vacation Fund", current: 250, target: 1000 },
  ];

  const budgets = [
    { id: 1, category: "Food", spent: 320, limit: 500 },
    { id: 2, category: "Transport", spent: 220, limit: 200 },
    { id: 3, category: "Shopping", spent: 90, limit: 150 },
  ];

  const alerts = [
    { id: 1, text: "You exceeded transport spending by 10%" },
    { id: 2, text: "Consider increasing savings this month" },
  ];

  const transactions = [
    { id: 1, desc: "Grocery Store", amount: -55, date: "2025-11-12" },
    { id: 2, desc: "Salary Deposit", amount: 3200, date: "2025-11-10" },
    { id: 3, desc: "Electricity Bill", amount: -90, date: "2025-11-09" },
  ];

  const formatCurrency = (n) =>
    n >= 0 ? `+$${n.toLocaleString()}` : `-$${Math.abs(n).toLocaleString()}`;

  const progressPct = (current, target) =>
    Math.min(100, Math.round((current / target) * 100));

  return (
    <div className="min-h-screen bg-gray-50 p-6">
      <div className="max-w-6xl mx-auto">
        {/* Header */}
        <header className="flex items-center justify-between mb-6">
          <div className="flex items-center gap-4">
            <div className="w-10 h-10 bg-indigo-600 rounded flex items-center justify-center text-white font-bold">FI</div>
            <h1 className="text-2xl font-semibold">Goal-Driven Dashboard</h1>
          </div>
          <div className="flex items-center gap-4">
            <div className="text-right">
              <div className="text-sm text-gray-500">Total Balance</div>
              <div className="text-lg font-bold">${balance.toLocaleString()}</div>
            </div>
            <button className="px-4 py-2 bg-indigo-600 text-white rounded shadow">Add Money</button>
          </div>
        </header>

        <main className="grid grid-cols-1 lg:grid-cols-3 gap-6">
          {/* Left column: Goals & Budgets (spans 2/3 on large screens) */}
          <section className="lg:col-span-2 space-y-6">
            {/* Savings Goals Card */}
            <div className="bg-white rounded-2xl p-5 shadow-sm">
              <div className="flex items-center justify-between mb-4">
                <h2 className="text-lg font-medium">Savings Goals</h2>
                <div className="text-sm text-gray-500">Progress toward targets</div>
              </div>

              <div className="space-y-4">
                {goals.map((g) => {
                  const pct = progressPct(g.current, g.target);
                  return (
                    <div key={g.id} className="flex flex-col gap-2">
                      <div className="flex justify-between items-baseline">
                        <div className="font-medium">{g.title}</div>
                        <div className="text-sm text-gray-600">${g.current}/{g.target} ({pct}%)</div>
                      </div>

                      <div className="w-full bg-gray-100 rounded-full h-3 overflow-hidden">
                        <motion.div
                          initial={{ width: 0 }}
                          animate={{ width: `${pct}%` }}
                          transition={{ duration: 0.8 }}
                          className={`h-3 rounded-full bg-gradient-to-r from-green-400 to-teal-500`}
                        />
                      </div>
                    </div>
                  );
                })}
              </div>
            </div>

            {/* Budget Overview Card */}
            <div className="bg-white rounded-2xl p-5 shadow-sm">
              <div className="flex items-center justify-between mb-4">
                <h2 className="text-lg font-medium">Budget Overview</h2>
                <div className="text-sm text-gray-500">Track monthly limits</div>
              </div>

              <div className="space-y-4">
                {budgets.map((b) => {
                  const pct = progressPct(b.spent, b.limit);
                  const over = b.spent > b.limit;
                  return (
                    <div key={b.id} className="flex items-center justify-between">
                      <div className="w-1/3">
                        <div className="font-medium">{b.category}</div>
                        <div className="text-sm text-gray-500">${b.spent}/${b.limit}</div>
                      </div>
                      <div className="w-2/3">
                        <div className="w-full bg-gray-100 rounded-full h-3 overflow-hidden">
                          <motion.div
                            initial={{ width: 0 }}
                            animate={{ width: `${pct}%` }}
                            transition={{ duration: 0.8 }}
                            className={
                              `h-3 rounded-full ${over ? 'bg-red-400' : 'bg-indigo-500'}`
                            }
                          />
                        </div>
                      </div>
                    </div>
                  );
                })}
              </div>
            </div>

            {/* Alerts & Recommendations */}
            <div className="bg-white rounded-2xl p-5 shadow-sm">
              <h2 className="text-lg font-medium mb-3">Alerts & Recommendations</h2>
              <ul className="space-y-2 text-sm text-gray-700">
                {alerts.map((a) => (
                  <li key={a.id} className="flex items-start gap-3">
                    <div className="mt-0.5 w-2 h-2 bg-yellow-400 rounded-full" />
                    <div>{a.text}</div>
                  </li>
                ))}
              </ul>
            </div>
          </section>

          {/* Right column: Recent Transactions (1/3 on large screens) */}
          <aside className="space-y-6">
            <div className="bg-white rounded-2xl p-5 shadow-sm">
              <div className="flex items-center justify-between mb-4">
                <h2 className="text-lg font-medium">Recent Transactions</h2>
                <div className="text-sm text-gray-500">Last 30 days</div>
              </div>

              <div className="divide-y divide-gray-100">
                {transactions.map((t) => (
                  <div key={t.id} className="py-3 flex items-center justify-between">
                    <div>
                      <div className="font-medium">{t.desc}</div>
                      <div className="text-xs text-gray-500">{t.date}</div>
                    </div>
                    <div className={`font-semibold ${t.amount < 0 ? 'text-red-500' : 'text-green-600'}`}>
                      {formatCurrency(t.amount)}
                    </div>
                  </div>
                ))}
              </div>

              <div className="mt-4 text-center">
                <button className="px-3 py-2 rounded-md border border-gray-200 text-sm">View all</button>
              </div>
            </div>

            {/* Compact Insights Card */}
            <div className="bg-white rounded-2xl p-5 shadow-sm">
              <h3 className="text-sm font-medium">Quick Insights</h3>
              <ul className="mt-3 text-sm text-gray-600 space-y-2">
                <li>Projected savings this month: <span className="font-semibold">$850</span></li>
                <li>Top spending category: <span className="font-semibold">Food</span></li>
              </ul>
            </div>
          </aside>
        </main>

        {/* Footer / small print */}
        <footer className="mt-8 text-xs text-gray-400 text-center">Designed with user goals in mind  Replace mock data with API responses</footer>
      </div>
    </div>
  );
}
```
